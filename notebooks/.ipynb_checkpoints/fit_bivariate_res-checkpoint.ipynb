```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
from scipy.stats import norm, multivariate_normal
from scipy.optimize import minimize_scalar
```

```python
# Loading in the data.
def LoadData(energy):
    
    filename = f'/home/llr/ilc/ritzmann/work/ana_files_ECAL_QGSP_BERT_conf6_e-__GeV_5kevt_build_masked/data_nhits_wenergy_{energy}GeV.txt'
    df = pd.read_csv(filename, header=0)
    
    return df
```

```python
#calculate the means, and the covariance matrices for all the different energies
Energies = [1, 2, 5, 8, 10, 20, 40, 60, 80, 100, 150]
#Energies = [1, 2, 5, 8, 10, 20]

means1 = []
means2 = []
sigmas1 = []
sigmas2 = []
rhos = []

for energy in Energies:
    df = LoadData(energy)
    mean = np.mean(df.to_numpy(), axis=0)
    cov = np.cov(df.to_numpy(), rowvar=0)
    sigma1 = np.sqrt(cov[0][0])
    sigma2 = np.sqrt(cov[1][1])
    #rho = np.sqrt(np.abs(cov[1][0]))
    rho = cov[1][0]/(sigma1*sigma2)
    #rho = np.abs(cov[1][0])
    means1.append(mean[0])
    means2.append(mean[1])
    sigmas1.append(sigma1)
    sigmas2.append(sigma2)
    rhos.append(rho)
    
df1 = pd.DataFrame({
    
    'energy': Energies,
    'mean_nhit':means1,
    'mean_sumE':means2,
    'sigma_nhit':sigmas1,
    'sigma_sumE':sigmas2,
    'rhoNew':rhos
})
df1
```

```python
def parabola(x, a, b, c):
    y = a*x*x + b*x + c
    return y

nhit_params, _ = curve_fit(f=parabola,
                          xdata = df1['mean_nhit'],
                          ydata = df1['energy'])
sumE_params, _ = curve_fit(f=parabola,
                          xdata = df1['mean_sumE'],
                          ydata = df1['energy'])
sigma_nhit_params, _ = curve_fit(f=parabola,
                          xdata = df1['sigma_nhit'],
                          ydata = df1['energy'])
sigma_sumE_params, _ = curve_fit(f=parabola,
                          xdata = df1['sigma_sumE'],
                          ydata = df1['energy'])
rhoNew_params, _ = curve_fit(f=parabola,
                          xdata = df1['rhoNew'],
                          ydata = df1['energy'])
```

```python
# Idea: Make a list of energies; calculate the bigaus parameters for all of them.
# Then take a random point (nhit, sum_E) and find the distribution with the greatest prob. density. that it could belong to.
energies = np.linspace(0.3, 23, 10000)

def inv_fn(x, params, y0): #INPUT: nhit_mean, or something, parameters of fit, and y0, the energy to be found.
    a = params[0]
    b = params[1]
    c = params[2]
    y = parabola(x, a, b, c)
    return y - y0

#root_scalar finds zero of a function
from scipy.optimize import root_scalar
nhits = []
sumE = []
sigma_nhits = []
sigma_sumE = []
rhonew = []
for energy in energies:
    nhits_sol = root_scalar(f=inv_fn, args=(nhit_params,energy),x0=0,x1=10)
    sumE_sol = root_scalar(f=inv_fn, args=(sumE_params,energy),x0=0,x1=1000)
    sigma_nhits_sol = root_scalar(f=inv_fn, args=(sigma_nhit_params, energy), x0=0, x1=10)
    sigma_sumE_sol = root_scalar(f=inv_fn, args=(sigma_sumE_params, energy), x0=0, x1=10)
    rhonew_sol = root_scalar(f=inv_fn, args=(rhoNew_params, energy), x0=0, x1=10)
    nhits.append(nhits_sol.root)
    sumE.append(sumE_sol.root)
    sigma_nhits.append(sigma_nhits_sol.root)
    sigma_sumE.append(sigma_sumE_sol.root)
    rhonew.append(rhonew_sol.root)
```

```python
# Create a bivariate normal distribution for each energy
normdist = []
for i in range(len(energies)):
    energy = energies[i] 
    mean = [nhits[i], sumE[i]]# this will be the mean.
    cov = np.eye(2)
    cov[0][0] = sigma_nhits[i]**2.0
    cov[1][1] = sigma_sumE[i]**2.0
    cov[1][0] = rhonew[i]*sigma_nhits[i]*sigma_sumE[i]
    cov[0][1] = rhonew[i]*sigma_nhits[i]*sigma_sumE[i]
    print(cov)
    # cov will be the covariance matrix.
    var = multivariate_normal(mean=mean, cov=cov, allow_singular=True)
    # var will be the multivariate normal distribution.
    normdist.append(var)
```

```python
def probability(x, y, normdist=normdist):
    point = [x, y]
    probs = []
    for i in range(len(normdist)):
        probs.append(normdist[i].pdf(point))
    return probs
probability = np.vectorize(probability)


for energy in Energies:
    df=df.append(LoadData(energy))
df
```

```python
Ntrials=100
res=[]
e=[]
Energies = [1, 2, 5, 8, 10, 20]
for energy in Energies:
    print(f"Energy(GeV):{energy}")
    df = LoadData(energy)
    trials = []
    for i in range(Ntrials):
        randint = np.random.randint(0,len(df))
        point = df.iloc[randint].to_numpy()
        prob = probability(point[0],point[1])
        trials.append(energies[np.argmax(prob)])
    trials = np.array(trials)
    print(f"After {Ntrials} trials,estimated energy:{trials.mean().round(3)}, with std: {trials.std().round(3)}, and resolution:{np.round(trials.std()/trials.mean(), 3)}")
    res.append(trials.std()/trials.mean())
    e.append(energy)
print(f"Lower Bound on Resolution:{(np.max(energies)-np.min(energies))/len(energies)} due to spacing")
df3 = pd.DataFrame({
    'energy': e,
    'resolution':res
    
})
df3.head()
```
