```python
sum_hits_corr_gauss = pd.read_csv('/home/llr/ilc/ritzmann/work/res_csv_files/conf6_e-_GeV_5kevt_-42_-42_build_masked_params_sum_hits_corr_gauss_test.txt', header=0)
sum_hits_corr_gauss2s = pd.read_csv('/home/llr/ilc/ritzmann/work/res_csv_files/conf6_e-_GeV_5kevt_-42_-42_build_masked_params_sum_hits_corr_gauss2s_test.txt', header=0)
sum_hits_corr_gamma = pd.read_csv('/home/llr/ilc/ritzmann/work/res_csv_files/conf6_e-_GeV_5kevt_-42_-42_build_masked_params_sum_hits_corr_gamma_test.txt', header=0)

plt.figure(figsize=(16,9))
plt.plot(np.sqrt(sum_hits_corr_gauss['energy']), sum_hits_corr_gauss['goodness'], 'o-', label = 'gauss')
plt.plot(np.sqrt(sum_hits_corr_gauss2s['energy']), sum_hits_corr_gauss2s['goodness'], 'o-', label = 'gauss 2$\sigma$')
plt.plot(np.sqrt(sum_hits_corr_gamma['energy']), sum_hits_corr_gamma['chisq'] / sum_hits_corr_gamma['ndf'], 'o-', label = 'gamma')
plt.legend(fontsize=30)
plt.grid()
plt.title('goodness-of-fit for number of hits', fontsize=15)
plt.ylabel('$\chi^2$ / NDF', fontsize = 15)
plt.xlabel('$\sqrt{E[GeV]}$', fontsize=15)
plt.margins(x=0.1, y=0.1)
```
