```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
shower = pd.read_csv("/home/llr/ilc/ritzmann/work/ECAL_QGSP_BERT_conf6_e-_GeV_5kevt_-42_-42_build_masked_shower_leakage.txt", header=0)
```

```python
plt.figure(figsize = (16,9))
plt.plot(shower['energy'], shower['shower_leakage'], 'o-', label = 'energy leakage')
plt.grid()
plt.legend()
plt.xlabel('energy [GeV]', fontsize=14)
plt.ylabel('energy leakage', fontsize=14)
plt.title('longitudinal energy leakage', fontsize=14)
plt.savefig('/home/llr/ilc/ritzmann/work/energy_leakage_e-.png')
```
