```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_sum_events = pd.read_csv("/home/llr/ilc/ritzmann/work/photon_analysis/ECAL_QGSP_BERT_TB2022-06_CONF6_gamma_GeV_params_sum_events.txt", header=0)
data_sum_energy = pd.read_csv("/home/llr/ilc/ritzmann/work/photon_analysis/ECAL_QGSP_BERT_TB2022-06_CONF6_gamma_GeV_params_sum_energy.txt", header=0)
data_sum_energy_weighted = pd.read_csv("/home/llr/ilc/ritzmann/work/photon_analysis/ECAL_QGSP_BERT_TB2022-06_CONF6_gamma_GeV_params_sum_energy_weighted.txt", header=0)
```

```python
plt.figure(figsize=(16,9))
plt.plot(1/np.sqrt(data_sum_events['energy']), data_sum_events['sigma']/data_sum_events['mean'], 'o-', label = 'sum hits')
plt.plot(1/np.sqrt(data_sum_energy['energy'])[1:], data_sum_energy['sigma'][1:]/data_sum_energy['mean'][1:], 'o-', label = 'sum energy')
plt.plot(1/np.sqrt(data_sum_energy_weighted['energy'])[1:], data_sum_energy_weighted['sigma'][1:]/data_sum_energy_weighted['mean'][1:], 'o-', label = 'weighted sum energy')
plt.grid()
plt.title('resolution vs. energy', fontsize = 17)
plt.xlabel('1/$\sqrt{E [GeV]}$', fontsize = 17)
plt.ylabel('$\sigma$/$\mu$', fontsize = 17)
plt.legend()
```
